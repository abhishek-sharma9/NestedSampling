```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

m_true = 0.75
c_true = 5.

# data generation
N = 100
x = np.sort(10 * np.random.rand(N))
signal = m_true * x + c_true

sigma = 0.1 + 0.5 * np.random.rand(N)  # errors or standard devitaion of each data point
noise = sigma*np.random.randn(N) # gaussian noise with 0 mean and standard deviation sigma

data = signal + noise # data 

f = open("data.txt", "w+")

for i in range(N):
    
    f.write("%f %f %f\n" %(x[i], data[i], sigma[i]))
    
f.close()
plt.figure(figsize = (12, 8))
plt.errorbar(x, data, yerr=sigma, fmt=".k", capsize=0)
x0 = np.linspace(0, 10, 1000)
plt.plot(x0, m_true * x0 + c_true, "k", alpha=0.3, lw=3)
plt.xlim(0, 10)
plt.xlabel("x", fontsize = "15")
plt.ylabel("y", fontsize = "15")
```
