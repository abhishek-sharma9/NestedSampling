# Beta Distribution: $B(x,\alpha,\beta)$

\begin{equation}
B(x,\alpha,\beta) = \frac{1}{B(\alpha, \beta)}x^{\alpha - 1}(1 - x)^{\beta - 1}, \quad 0 \leq x \leq 1
\end{equation}

\begin{equation}
B(x,\alpha,\beta) = \frac{\Gamma(\alpha+\beta)}{\Gamma (\alpha)\Gamma (\beta)}x^{\alpha - 1}(1 - x)^{\beta - 1}, \quad 0 \leq x \leq 1
\end{equation}

Here, we are interested in the case where $\alpha = N_l$, the number of live points and $\beta = 1$. Therefore, 
\begin{equation}
B(x,N_l, 1) = \frac{\Gamma(N_l+1)}{\Gamma (N_l)\Gamma (1)} x^{N_l-1} = N_l x^{N_l-1}
\end{equation}


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
x = np.linspace(0, 1, 1000)
```

```python
N = [5, 10, 15, 20, 50, 100]

# fig, ax = plt.subplots(2, 3)
plt.figure(figsize = (15, 10))
for i in range(len(N)):
    
    plt.subplot(2, 3, i+1)
    plt.plot(x, st.beta.pdf(x, N[i], 1), 'k')
    plt.xlabel('x', fontsize = '12')
    plt.ylabel('p(x)', fontsize = '12')
    plt.grid()
    plt.title(f'N = {N[i]}')
    
```

```python
plt.figure(figsize = (16, 10))
for i in range(len(N)):
    
    plt.subplot(2, 3, i+1)
    plt.plot(x, st.beta.logpdf(x, N[i], 1), 'k')
    plt.xlabel('x', fontsize = '12')
    plt.ylabel('p(x)', fontsize = '12')
    plt.grid()
    plt.title(f'N = {N[i]}')
```

Suppose we sample $N_l$ points uniformly between 0 and 1. i.e,  $s \sim U[0,1]$. Then, the random variable: "maximum value out of the sampled points" obey $B(N_l,1)$ distribution.

$$B(x, N_l, 1) = N_l x^{N_l-1}$$

```python
n = 5
maxima = []
i = 0
while i < 1000:                               # sampling "n" points 1000 times.
    maxima.append(max(np.random.rand(n)))
    i += 1
```

```python
plt.figure(figsize = (11,7))
plt.hist(maxima, bins=25, density = True)
plt.plot(x, st.beta.pdf(x, n, 1), 'k', label = 'beta distribution')
plt.xlabel('x')
plt.ylabel('No. of sampled points' )
plt.legend()
plt.show()
```
